```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('./Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
cols_to_use = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount',
               'Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
```

```python
dataset = dataset[cols_to_use]
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount','Distance','Bedroom2','Bathroom','Car']
dataset[cols_to_fill_zero] = dataset[cold_to_fill_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset["Landsize"].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset["BuildingArea"].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
X = dataset.drop('Price', axis=1)
y=dataset['Price']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.3, random_state=2)
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
reg.score(X_train, y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=50, max_iter=1000, tol=0.01)
lasso_reg.fit(X_train, y_train)
lasso_reg.score(X_test, y_test)
```

```python
lasso_reg.score(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=50, max_iter=1000, tol=0.01)
ridge_reg.fit(X_train, y_train)
ridge_reg.score(X_test, y_test)
```

```python
ridge_reg.score(X_train, y_train)
```
