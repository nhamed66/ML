```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('HR_comma_sep.csv')
left = df[df.left==1]
retained = df[df.left==0]
df.groupby('left').mean()
salary_cross = pd.crosstab(df.salary, df.left)
print(salary_cross.head())
left_ratio_salary = salary_cross.loc[:,1]/salary_cross.loc[:,0]
plt.bar(['high','low','medium'], left_ratio_salary)
plt.show()
```

```python
department_cross = pd.crosstab(df.Department, df.left)
print(department_cross.head())
left_ratio_department  = department_cross.loc[:,1]/department_cross.loc[:,0]
left_ratio_department = left_ratio_department.sort_values()
print(left_ratio_department)
plt.figure(figsize=(10,6))
plt.bar(['management','RandD','product_mng','IT','matketting','sales','support', 'technical','accounting','hr'],left_ratio_department)
```
