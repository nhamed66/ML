```python
import pandas as pd
df = pd.read_csv('heart_2.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df1 = df[df.Cholesterol<= df.Cholesterol.mean()+3*df.Cholesterol.std()]
df1.shape
```

```python
df2 = df1[df1.Cholesterol>= df.Cholesterol.mean()-3*df.Cholesterol.std()]
df2.shape
```

```python
df3 = df2[df2.RestingBP<= df.RestingBP.mean()+3*df.RestingBP.std()]
df3.shape
```

```python
df4 = df3[df3.RestingBP>= df.RestingBP.mean()-3*df.RestingBP.std()]
df4.shape
```

```python
df5 = df4[df4.Age>= df.Age.mean()-3*df.Age.std()]
df5.shape
```

```python
df6 = df5[df5.Age<= df.Age.mean()+3*df.Age.std()]
df6.shape
```

```python
df7 = df6[df6.FastingBS<= df.FastingBS.mean()+3*df.FastingBS.std()]
df7.shape
```

```python
df8 = df7[df7.FastingBS>= df.FastingBS.mean()-3*df.FastingBS.std()]
df8.shape
```

```python
df9 = df8[df8.MaxHR>= df.MaxHR.mean()-3*df.MaxHR.std()]
df9.shape
```

```python
df10 = df9[df9.Oldpeak>= df.Oldpeak.mean()-3*df.Oldpeak.std()]
df11 = df10[df10.Oldpeak<= df.Oldpeak.mean()+3*df.Oldpeak.std()]
df11.shape
```

```python
df11.ChestPainType.unique()
```

```python
df11.RestingECG.unique()
```

```python
df11.ST_Slope.unique()
```

```python
df11.Sex.unique()
```

```python
df11.ExerciseAngina.unique()
```

```python
df12 = df11.copy()
df12.ExerciseAngina.replace(
    {
        'N':0,
        'Y':1
    },
    inplace=True)
```

```python
df12.RestingECG.replace(
    {
        'Normal':1,
        'ST':2,
        'LVH':3
    },
    inplace=True)
df12.ST_Slope.replace(
    {
        'Up':1,
        'Flat':2,
        'Down':3
    },
    inplace=True)
df12.head()
```

```python
df13 = pd.get_dummies(df12, drop_first=True)
df13.head()
```

```python
X=df13.drop('HeartDisease', axis='columns')
y=df13.HeartDisease


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=20)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
scores = cross_val_score(SVC(),X,y, cv=5)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(
    base_estimator=SVC(),
    n_estimators=100,
    max_samples=0.8,
    random_state=0
)
scores = cross_val_score(bag_model,X,y,cv=5)
scores.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier

scores = cross_val_score(DecisionTreeClassifier(random_state=0),X,y,cv=5)
scores.mean()
```

```python
bag_model = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(random_state=0),
    n_estimators=100,
    max_samples=0.9,
    oob_score=True,
    random_state=0
)
scores = cross_val_score(bag_model,X,y,cv=5)
scores.mean()
```
